## **เลือกแค่ค่าล่าสุด หากเกิด Duplicate Row**

 ~

````python
value = ['Systolic BP', 'Diastolic BP', 'BMI', 'BSA']
filtered = vital1[vital1['ResultItemName'].isin(value)]

grouped_check = filtered.groupby(['PatientUID', 'PatientVisitUID', 'ResultItemName']).agg({'ResultValue': 'last'}).reset_index()
grouped_check
````

````python
import pandas as pd
import numpy as np

non_numeric_values = grouped_check[~grouped_check['ResultValue'].apply(lambda x: isinstance(x, (int, float)) or str(x).replace('.', '', 1).isdigit())]
print("Non-numeric values in ResultValue column:")
print(non_numeric_values)
grouped_check['ResultValue'] = pd.to_numeric(grouped_check['ResultValue'], errors='coerce')

# Pivot the DataFrame
pivot_df = grouped_check.pivot_table(index=['PatientUID', 'PatientVisitUID'], columns='ResultItemName', values='ResultValue').reset_index()
pivot_df.columns.name = None
pivot_df
````
