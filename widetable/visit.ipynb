## เก็บแค่ visit ที่มีการตรวจ eGFR

```python
GFR = pd.read_csv('/content/drive/MyDrive/kidney/GFR.csv')
GFR.drop(columns='Unnamed: 0')
```

```python
patient_visit_list = GFR['PatientVisitUID'].unique().tolist()
visit = selected[selected['PatientvisitUID'].isin(patient_visit_list)]
visit.head()
```

```python
import pandas as pd

df = pd.DataFrame(visit)
df['InPatient'] = df['ENTY'].apply(lambda x: 1 if x == 'InPatient' else 0)
df['OutPatient'] = df['ENTY'].apply(lambda x: 1 if x == 'Out Patient' else 0)
df['Emergency'] = df['ENTY'].apply(lambda x: 1 if x == 'E' else 0)
df
```
